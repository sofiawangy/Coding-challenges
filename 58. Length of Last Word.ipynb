```python
def lengthOfLastWord(s):
    
    ls = len(s)
    slow = -1
    
    while slow >= -ls and s[slow] == ' ':
        slow -= 1
        
    fast = slow
    
    while fast >= -ls and s[fast] != ' ':
        fast -= 1
    
    return slow - fast
```

```python
s = "a "
lengthOfLastWord(s)
```
