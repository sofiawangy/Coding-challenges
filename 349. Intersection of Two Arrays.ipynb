```python
# Input: nums1 = [1,2,2,1], nums2 = [2,2]
# Output: [2]
def intersection(nums1, nums2):
    """
    :type nums1: List[int]
    :type nums2: List[int]
    :rtype: List[int]
    """
    return list(set(nums2) & set(nums1))
```

```python


nums1 = [4,7,9,7,6,7]
nums2 = [5,0,0,6,1,6,2,2,4]
intersection(nums1, nums2)
```
