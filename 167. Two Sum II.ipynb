```python
nums = [-1,0]
target = -1
```

```python
def twoSum(numbers, target):
    """
    :type numbers: List[int]
    :type target: int
    :rtype: List[int]
    """
    dic = {}
    for i, n in enumerate(numbers,1):
        diff = target - n
        if diff in dic:
            return [dic[diff], i]
        dic[n] = i
```

```python
twoSum(nums, target)
```
