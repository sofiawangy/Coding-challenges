```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

a = b = ListNode(0)

print(a.next, b.next)
```

### separate assignment -- function / class

```python
b = ListNode(0)
a = b

print(a.next, b.next)

b.next = 'updated'

print(a.next, b.next)
```

### adding DEEP COPY: separate assignment -- function / class

```python
import copy

b = ListNode(0)
a = copy.copy(b)

print(a.next, b.next)

b.next = 'updated'

print(a.next, b.next)
```

```python
class Solution:
    def nextPermutation(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """

        if not nums:
            return []

        # testcase: '[3,2,1]'
        i = len(nums) - 1  # i=2

        while i > 0:
            if nums[i] <= nums[i-1]:
                i -= 1 # i=1, i=0
            else:
                break

        if not i:
            return nums[::-1]

        loc1 = i - 1

        min_bigger = float('inf')

        for j, v in enumerate(nums[i:]):
            if v > nums[loc1] and v < min_bigger:
                min_bigger = v
                loc2 = i+j

        nums[loc1], nums[loc2] = nums[loc2], nums[loc1]
        nums[i:] = sorted(nums[i:])

        return nums
```

- However, if we created DEEP COPIES of the input links ahead of func execution, the deep copies stayed the same
- We believe this is the way our algorithem is designed to do (minimizing memory consumption, but modifying things "in place")
