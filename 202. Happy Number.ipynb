```python
def isHappy(n):
    """
    :type n: int
    :rtype: bool
    """
    a = set()
    
    def get_next(n):
        cumm = 0
        
        while n > 0:
            digit, r = divmod(n, 10)
            cumm += r**2
            n = digit
        
        return cumm
    
    while n != 1 and n not in a:
        a.add(n)
        n = get_next(n)
    
    return n == 1
```

```python
n = 19
isHappy(n)
```
