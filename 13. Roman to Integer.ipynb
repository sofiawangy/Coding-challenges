```python
def romanToInt(s):
    value = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    prev = None
    result = 0
    
    for i in s:
        curr = value[i]
        
        result += curr
        
        if prev and curr > prev:
            result -= 2* prev
            
        prev = curr
    
    return result
```

```python
romanToInt("MCMXCIV")
```
