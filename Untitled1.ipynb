```python
# Input: -123
# Output: -321
def reverse(x):
    """
    :type x: int
    :rtype: int
    """
    if x < 0:
        sign = -1
    else:
        sign = 1
    
    x = abs(x)
    rev = 0
        
    while x > 0:
        x, dig = divmod(x,  10)
        rev = rev*10 + dig
        print(x)
    
    return rev*sign
```
