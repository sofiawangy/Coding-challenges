```python
def removeDuplicates(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    if len(nums) == 0: return 0
    
    i=0
    
    for j in range(1,len(nums)):
        print(nums[j], nums[i], i)
        
        if nums[j] != nums[i]:
            print(nums[j], nums[i], i)
            i+=1
            nums[i] = nums[j]
            
    
    return i+1
```

```python
nums = [0,0,1,1,1,2,2,3,3,4]

i=0

for j in range(1,len(nums)):
    print(nums[j], j)

    if nums[j] != nums[i]:
#         print(nums[j], nums[i], i)
        i+=1
        nums[i] = nums[j]
    
    print(nums)
```
