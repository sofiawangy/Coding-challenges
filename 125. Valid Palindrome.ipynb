```python


def isPalindrome(s):

    s = ''.join(i for i in s if i.isalnum()).lower()
    l = len(s)

    for i in range(l//2):
        if s[i] != s[l-i-1]:
            return False
        
    return True
```

```python
s = "race a car"
isPalindrome(s)
# for i in range(len(s)//2):
#     print(i, s[i])
```
