```python
# load the csv as a dataframe

# loop through the 2nd col (containing all the urls)
# for each url, we try to get the author name


# the whole purpose is to see whether we can get info through CML, much faster than Selenium.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('news.csv', header=None)
```

```python
urls = list(df.iloc[:, 1])
```

## urls is ready, now looping through

```python
def get_author(url):
    
    response = requests.get(url)
    author = re.findall(r'(?<="Person","name":")[^.\s]*',response.text)
    if response.status_code == 404:
        pass
    response.close() # important!
    # nice to add a 'pause' everytime -- this can (should) be done outside the function
    
    return author

# requests.get('https://www.cnbc.com/2019/07/24/bristol-myers-releases-mixed-opdivo-lung-cancer-results.html')
```

```python
author_list = []

for url in urls:
    author = get_author(url)
    print(url)
    author_list.append(author[0])
```
