```python
c = [[259,770],[448,54],[926,667],[184,139],[840,118],[577,469]]
```

```python
def twoCitySchedCost(costs):
    costs.sort(key = lambda x: x[0] - x[1])
    
    n = len(costs)//2
    
    res = 0
    
    for i in range(n):
        res += costs[i][0] + costs[i+n][1]
    
    return res
    
twoCitySchedCost(c)
```
