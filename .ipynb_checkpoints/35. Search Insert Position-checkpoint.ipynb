```python
def searchInsert(nums, target):
    
    left, right = 0, len(nums)-1
    
    while left <= right:
        middle = (left + right)//2
        
        if nums[middle] == target:
            return target
        elif target < nums[middle]:
            right = middle - 1
        else:
            left = middle + 1
        
        return left
```

```python
nums = [1,3,5,6]
target = 0
searchInsert(nums, target)
```
