```python
# Count the number of prime numbers less than a non-negative number, n.

# Example:

# Input: 10
# Output: 4
# Explanation: There are 4 prime numbers less than 10, they are 2, 3, 5, 7.
    

def countPrimes(n):
    """
    :type n: int
    :rtype: int
    """
    if n <= 2:
        return 0

    isPrime = [True]*n
    isPrime[0] = isPrime[1] = False

    i = 2

    #cut down iteration from n to n**1/2
    while i*i < n:
        if isPrime[i]:
            
            j = i*i
            while j < n:
                isPrime[j] = False
                j+= i

        i+=1

    return sum(isPrime)
```
