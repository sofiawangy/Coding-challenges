```python
# test
l = [4,1,2,1,2]

while l:
    a = l[0]
    
    if a not in l[1:]:
        print(a)
        break
    
    else:
        l = l[1:]
        l.remove(a)
```

```python
#submission
class Solution(object):

    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        while nums:
            a = nums[0]

            if a not in nums[1:]:
                return a

            else:
                nums = nums[1:]
                nums.remove(a)
```
