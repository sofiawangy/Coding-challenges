```python
def intersect(nums1, nums2):
    """
    :type nums1: List[int]
    :type nums2: List[int]
    :rtype: List[int]
    """

    counter = {}
    res = []
    for i in nums1:
        if i not in counter:
            counter[i] = 1
        else:
            counter[i] += 1
    
    for i in nums2:
        if i in counter and counter[i] > 0:
            res.append(i)
            counter[i] -= 1
    
    return res
```

```python
nums1 = [4,7,9,7,6,7]
nums2 = [5,0,0,6,1,6,2,2,4]

intersect(nums1, nums2)
```
