```python
def maxSubArray(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    curr = greatest = nums[0]
    
    for i in nums[1:]:
        curr = max(i, curr + i)
        greatest = max(greatest, curr)
        
    return greatest
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
maxSubArray(nums)
```
