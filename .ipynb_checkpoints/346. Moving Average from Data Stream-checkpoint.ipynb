```python
class MovingAverage(object):

    def __init__(self, size):
        """
        Initialize your data structure here.
        :type size: int
        """
        self.size = size
        self.slots = []
        self.count = 0

    def next(self, val):
        """
        :type val: int
        :rtype: float
        """

        self.count += 1
        if self.count > self.size:
            self.slots.pop(0)
            self.count = self.size
        
        self.slots.append(val)

        return sum(self.slots)/self.count
```
