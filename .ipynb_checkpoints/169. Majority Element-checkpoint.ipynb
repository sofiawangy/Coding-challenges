```python
def majorityElement(nums):
    
    count = 0
    candidate = 0
    
    for n in nums:
        
        if count == 0:
            candidate = n
        
        if n == candidate:
            count+=1
        else:
            count-=1
    
    return candidate
```

```python
nums = [2,2,1,1,1,2,2]

majorityElement(nums)
```
