```python
def isValid(s):
    """
    :type s: str
    :rtype: bool
    """
    if not s:
        return True
    
    close = {')': '(', 
             ']': '[',
             '}': '{'}

    l1 = []

    for i in s:
        
        if i in close:
            top = l1.pop() if l1 else '0'
            if close[i] != top:
                return False
        else:
            l1.append(i)
    
    return not l1
```

```python
s1 = '[]]'
isValid(s1)
```
