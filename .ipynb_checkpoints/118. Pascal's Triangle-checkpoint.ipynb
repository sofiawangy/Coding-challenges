```python
def generate(numRows):
    """
    :type numRows: int
    :rtype: List[List[int]]
    """
    triangle = []
    for i in range(numRows):
        row = [1]*(i+1)
        print(row)
        for j in range(1, len(row)-1):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
            print(j)
            
        triangle.append(row)

    return triangle
```

```python
generate(4)
```
