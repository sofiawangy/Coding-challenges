```python
def plusOne(self, digits):
    last = len(digits) - 1

    if digits[last] < 9:
        digits[last] = digits[last] +1
        return digits
    else:
        digits[last] = 0
        digits[0:last] = self.plusOne(digits[0:last])
        return digits
```
