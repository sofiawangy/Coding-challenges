```python
# Input: nums = [1,2,3,1,2,3], k = 2
# Output: false
def containsNearbyDuplicate(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: bool
    """
    counter = {}
    
    for i, num in enumerate(nums):
        if num not in counter:
            counter[num] = i
        else:
            if i - counter[num] <= k:
                return True
            else:
                counter[num] = i
    
    return False
```

```python
nums = [1,2,3,1]
k = 3
containsNearbyDuplicate(nums, k)
```
