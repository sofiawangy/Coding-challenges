```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x < 0:
            sign = -1
        else:
            sign = 1

        rev = 0

        x = abs(x)

        while x:
            a = x % 10
            rev = rev*10 + a
            x = int(x/10)

        return 0 if rev > 2** 31 else rev*sign
```
