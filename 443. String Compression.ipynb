```python
def compress(chars):
    curr = '#'
    count = 1
    start = 0
    
    
    
    for i in range(len(chars)):
#         start = i
        
        if chars[i] == curr:
            count += 1
            if count < 2:
                start += 1
        else:
            start = i
            if count > 1:
                chars[start] = count

            curr = chars[i]
            count = 1

        print(chars[i], count, start)
        
    return chars
```
