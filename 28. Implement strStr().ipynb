```python
def strStr(haystack, needle):
    if len(needle) == 0:
        return 0
    
    if needle not in haystack:
        return -1
    
    for i in range(0, len(haystack)):
        if haystack[i:i+len(needle)] == needle:
            return i
```

```python
haystack = "heoll"
needle = "ll"

strStr(haystack, needle)
# haystack[0:2]
```
