```python
# s = "leetcode"

def firstUniqChar(s):
    """
    :type s: str
    :rtype: int
    """
    d = {}

    for i in s:
        if i not in d.keys():
            d[i] = 1
        else:
            d[i] += 1

    for i, c in enumerate(d.values()):
        if c == 1:
            return i


    return -1
            
firstUniqChar(s2)
```
